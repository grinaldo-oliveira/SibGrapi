```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Diretórios onde estão os embeddings
#directories = ["Data/terumo-data-jpeg", "Data/terumo-val"]
directories = ["Data/terumo-data-jpeg"]

# Cria uma lista de todos os arquivos .npy nos diretórios
npy_files = []
for dir_path in directories:
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.npy'):
            npy_files.append(os.path.join(dir_path, file_name))

# Função para carregar dados
def load_embbed(pickle_path: str) -> tuple:
    data = np.load(pickle_path, allow_pickle=True).tolist()
    file_refs = data['fnames']
    embeds = data['embeddings']
    labels = data['labels']
    classes = data['classes']
    embeds = np.vstack(embeds)
    labels = np.hstack(labels)
    return file_refs, embeds, labels, classes

# Função para treinar modelo com validação cruzada e gerar métricas
def train_and_evaluate(X, y, class_names):
    # Normalizar os dados
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Criar o modelo com camada escondida de 2049 neurônios
    model = MLPClassifier(
        hidden_layer_sizes=(2049,),  # Uma camada escondida com 2049 neurônios
        max_iter=1000,
        activation='relu',
        solver='adam',
        random_state=42
    )

    # Configurar validação cruzada com 5 folds
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    # Armazenar métricas para cada fold
    fold_accuracies = []
    fold_precisions = []
    fold_recalls = []
    fold_f1s = []
    fold_cms = []

    # Executar validação cruzada manualmente para capturar mais métricas
    fold_idx = 1
    for train_index, test_index in kf.split(X_scaled):
        X_train, X_test = X_scaled[train_index], X_scaled[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Treinar o modelo
        model.fit(X_train, y_train)

        # Fazer previsões
        y_pred = model.predict(X_test)

        # Calcular métricas
        accuracy = accuracy_score(y_test, y_pred)

        # Para multi-classe, calcular média ponderada
        precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        cm = confusion_matrix(y_test, y_pred)

        # Armazenar métricas
        fold_accuracies.append(accuracy)
        fold_precisions.append(precision)
        fold_recalls.append(recall)
        fold_f1s.append(f1)
        fold_cms.append(cm)

        print(f"Fold {fold_idx}:")
        print(f"  Acurácia: {accuracy:.4f}")
        print(f"  Precisão: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-Score: {f1:.4f}")
        print(classification_report(y_test, y_pred, target_names=class_names))
        print("-" * 50)

        fold_idx += 1

    # Calcular médias
    avg_accuracy = np.mean(fold_accuracies)
    avg_precision = np.mean(fold_precisions)
    avg_recall = np.mean(fold_recalls)
    avg_f1 = np.mean(fold_f1s)

    # Somar matrizes de confusão
    if fold_cms:
        sum_cm = fold_cms[0]
        for cm in fold_cms[1:]:
            sum_cm += cm
    else:
        sum_cm = np.array([])

    return {
        'accuracies': fold_accuracies,
        'precisions': fold_precisions,
        'recalls': fold_recalls,
        'f1_scores': fold_f1s,
        'avg_accuracy': avg_accuracy,
        'avg_precision': avg_precision,
        'avg_recall': avg_recall,
        'avg_f1': avg_f1,
        'confusion_matrix': sum_cm
    }

# Função para plotar resultados
def plot_results(results_dict, filename):
    plt.figure(figsize=(15, 10))

    # Plot das métricas por fold
    plt.subplot(2, 2, 1)
    folds = range(1, 6)
    plt.plot(folds, results_dict['accuracies'], 'o-', label='Acurácia')
    plt.plot(folds, results_dict['precisions'], 's-', label='Precisão')
    plt.plot(folds, results_dict['recalls'], '^-', label='Recall')
    plt.plot(folds, results_dict['f1_scores'], 'd-', label='F1-Score')
    plt.xlabel('Fold')
    plt.ylabel('Valor')
    plt.title(f'Métricas por Fold - {os.path.basename(filename)}')
    plt.grid(True)
    plt.legend()

    # Barplot das médias das métricas
    plt.subplot(2, 2, 2)
    metrics = ['Acurácia', 'Precisão', 'Recall', 'F1-Score']
    values = [
        results_dict['avg_accuracy'],
        results_dict['avg_precision'],
        results_dict['avg_recall'],
        results_dict['avg_f1']
    ]
    plt.bar(metrics, values, color=['blue', 'green', 'red', 'purple'])
    plt.ylim(0, 1.0)
    for i, v in enumerate(values):
        plt.text(i, v + 0.02, f'{v:.4f}', ha='center')
    plt.title(f'Médias das Métricas - {os.path.basename(filename)}')

    # Matriz de confusão
    plt.subplot(2, 2, 3)
    if results_dict['confusion_matrix'].size > 0:
        sns.heatmap(
            results_dict['confusion_matrix'],
            annot=True,
            fmt='d',
            cmap='Blues',
            cbar=True
        )
        plt.title(f'Matriz de Confusão Acumulada - {os.path.basename(filename)}')
        plt.xlabel('Predito')
        plt.ylabel('Real')

    # Variabilidade das métricas (boxplot)
    plt.subplot(2, 2, 4)
    data_to_plot = [
        results_dict['accuracies'],
        results_dict['precisions'],
        results_dict['recalls'],
        results_dict['f1_scores']
    ]
    plt.boxplot(data_to_plot, labels=metrics)
    plt.ylim(0, 1.0)
    plt.title(f'Variabilidade das Métricas - {os.path.basename(filename)}')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()

    # Salvar a figura com nome do arquivo
    output_filename = f"resultados_{os.path.basename(filename).replace('.npy', '')}.png"
    plt.savefig(output_filename)
    print(f"Gráfico salvo como: {output_filename}")

    return plt

# Processar cada arquivo NPY
all_results = {}

for npy_file in npy_files:
    print(f"\nProcessando: {npy_file}")
    print("=" * 70)

    # Carregar dados
    file_refs, embeds, labels, classes = load_embbed(npy_file)

    # Verificar se os dados foram carregados corretamente
    if len(embeds) == 0 or len(labels) == 0:
        print(f"Dados insuficientes no arquivo {npy_file}. Pulando.")
        continue

    print(f"Número de exemplos: {len(embeds)}")
    print(f"Dimensão dos embeddings: {embeds.shape[1]}")
    print(f"Classes disponíveis: {classes}")

    # Verificar distribuição de classes
    unique_labels, counts = np.unique(labels, return_counts=True)
    print("\nDistribuição de classes:")
    for lbl, cnt in zip(unique_labels, counts):
        class_name = classes[lbl] if lbl < len(classes) else f"Classe {lbl}"
        print(f"  {class_name}: {cnt} exemplos")

    # Treinar e avaliar o modelo
    results = train_and_evaluate(embeds, labels, classes)
    all_results[npy_file] = results

    # Mostrar resultados finais
    print("\nResultados finais:")
    print(f"Acurácia média: {results['avg_accuracy']:.4f}")
    print(f"Precisão média: {results['avg_precision']:.4f}")
    print(f"Recall médio: {results['avg_recall']:.4f}")
    print(f"F1-Score médio: {results['avg_f1']:.4f}")

    # Plotar resultados
    plot = plot_results(results, npy_file)
    plt.show()  # Mostrar o gráfico

    print("\n" + "=" * 70)

# Comparação entre os diferentes arquivos (se houver mais de um)
if len(all_results) > 1:
    plt.figure(figsize=(12, 8))

    # Comparar acurácias
    file_names = [os.path.basename(f).replace('.npy', '') for f in all_results.keys()]
    accuracies = [results['avg_accuracy'] for results in all_results.values()]
    precisions = [results['avg_precision'] for results in all_results.values()]
    recalls = [results['avg_recall'] for results in all_results.values()]
    f1_scores = [results['avg_f1'] for results in all_results.values()]

    x = np.arange(len(file_names))
    width = 0.2

    plt.bar(x - 1.5*width, accuracies, width, label='Acurácia', color='blue')
    plt.bar(x - 0.5*width, precisions, width, label='Precisão', color='green')
    plt.bar(x + 0.5*width, recalls, width, label='Recall', color='red')
    plt.bar(x + 1.5*width, f1_scores, width, label='F1-Score', color='purple')

    plt.xlabel('Arquivos')
    plt.ylabel('Valor Médio')
    plt.title('Comparação de Desempenho entre Arquivos')
    plt.xticks(x, file_names, rotation=45, ha='right')
    plt.ylim(0, 1.0)
    plt.legend()
    plt.tight_layout()

    # Salvar gráfico comparativo
    plt.savefig("comparacao_entre_arquivos.png")
    print("Gráfico comparativo salvo como: comparacao_entre_arquivos.png")
    plt.show()

print("Processamento concluído para todos os arquivos.")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Diretórios onde estão os embeddings
#directories = ["Data/terumo-data-jpeg", "Data/terumo-val"]
directories = ["Data/terumo-data-jpeg"]

# Cria uma lista de todos os arquivos .npy nos diretórios
npy_files = []
for dir_path in directories:
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.npy'):
            npy_files.append(os.path.join(dir_path, file_name))

# Função para carregar dados
def load_embbed(pickle_path: str) -> tuple:
    data = np.load(pickle_path, allow_pickle=True).tolist()
    file_refs = data['fnames']
    embeds = data['embeddings']
    labels = data['labels']
    classes = data['classes']
    embeds = np.vstack(embeds)
    labels = np.hstack(labels)
    return file_refs, embeds, labels, classes

# Função para treinar modelo com validação cruzada e gerar métricas
def balance_samples(X, y):
    """
    Redimensiona todas as classes para terem o mesmo número de amostras que a menor classe.
    """
    # Identificar o número de amostras na menor classe
    unique_labels, counts = np.unique(y, return_counts=True)
    min_samples = np.min(counts)

    # Índices balanceados para cada classe
    balanced_indices = []

    for label in unique_labels:
        # Obter índices de todas as amostras desta classe
        indices = np.where(y == label)[0]

        # Se houver mais amostras que o mínimo, selecionar aleatoriamente
        if len(indices) > min_samples:
            np.random.seed(42)  # Para reprodutibilidade
            selected_indices = np.random.choice(indices, min_samples, replace=False)
            balanced_indices.extend(selected_indices)
        else:
            # Se for a menor classe, usar todas as amostras
            balanced_indices.extend(indices)

    # Converter para array numpy
    balanced_indices = np.array(balanced_indices)

    # Retornar dados balanceados
    return X[balanced_indices], y[balanced_indices]

def train_and_evaluate(X, y, class_names):
    # Balancear as amostras
    X_balanced, y_balanced = balance_samples(X, y)
    print(f"Dados originais: {X.shape[0]} amostras")
    print(f"Dados balanceados: {X_balanced.shape[0]} amostras")

    # Exibir contagem por classe após balanceamento
    unique_labels, counts = np.unique(y_balanced, return_counts=True)
    print("Distribuição após balanceamento:")
    for lbl, cnt in zip(unique_labels, counts):
        class_name = class_names[lbl] if lbl < len(class_names) else f"Classe {lbl}"
        print(f"  {class_name}: {cnt} exemplos")

    # Normalizar os dados
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_balanced)

    # Criar o modelo com camada escondida de 2049 neurônios
    model = MLPClassifier(
        hidden_layer_sizes=(2049,),  # Uma camada escondida com 2049 neurônios
        max_iter=2000,               # Aumentando o número máximo de iterações para garantir convergência
        activation='relu',
        solver='adam',
        alpha=0.0001,                # Regularização L2
        learning_rate_init=0.001,    # Taxa de aprendizado inicial
        early_stopping=True,         # Parar treinamento se o desempenho não melhorar
        validation_fraction=0.1,     # Fração dos dados de treino usados para validação
        n_iter_no_change=10,         # Número de iterações sem melhoria para considerar early stopping
        random_state=42
    )

    # Configurar validação cruzada com 5 folds
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    # Armazenar métricas para cada fold
    fold_accuracies = []
    fold_precisions = []
    fold_recalls = []
    fold_f1s = []
    fold_cms = []

    # Executar validação cruzada manualmente para capturar mais métricas
    fold_idx = 1
    for train_index, test_index in kf.split(X_scaled):
        X_train, X_test = X_scaled[train_index], X_scaled[test_index]
        y_train, y_test = y_balanced[train_index], y_balanced[test_index]

        # Treinar o modelo
        model.fit(X_train, y_train)

        # Fazer previsões
        y_pred = model.predict(X_test)

        # Calcular métricas
        accuracy = accuracy_score(y_test, y_pred)

        # Para multi-classe, calcular média ponderada
        precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        cm = confusion_matrix(y_test, y_pred)

        # Armazenar métricas
        fold_accuracies.append(accuracy)
        fold_precisions.append(precision)
        fold_recalls.append(recall)
        fold_f1s.append(f1)
        fold_cms.append(cm)

        print(f"Fold {fold_idx}:")
        print(f"  Acurácia: {accuracy:.4f}")
        print(f"  Precisão: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-Score: {f1:.4f}")

        # Corrigir o problema: usar apenas os nomes das classes que realmente aparecem nos dados
        # Determinar quais classes estão presentes nos dados de teste
        present_classes = np.unique(np.concatenate([y_test, y_pred]))
        # Filtrar os nomes das classes para incluir apenas as presentes
        filtered_class_names = [class_names[i] for i in present_classes if i < len(class_names)]

        # Se o número de classes únicas não corresponder ao tamanho de filtered_class_names, não use target_names
        if len(present_classes) == len(filtered_class_names):
            print(classification_report(y_test, y_pred, target_names=filtered_class_names))
        else:
            print(classification_report(y_test, y_pred))

        print("-" * 50)

        fold_idx += 1

    # Calcular médias
    avg_accuracy = np.mean(fold_accuracies)
    avg_precision = np.mean(fold_precisions)
    avg_recall = np.mean(fold_recalls)
    avg_f1 = np.mean(fold_f1s)

    # Somar matrizes de confusão
    if fold_cms:
        sum_cm = fold_cms[0]
        for cm in fold_cms[1:]:
            sum_cm += cm
    else:
        sum_cm = np.array([])

    return {
        'accuracies': fold_accuracies,
        'precisions': fold_precisions,
        'recalls': fold_recalls,
        'f1_scores': fold_f1s,
        'avg_accuracy': avg_accuracy,
        'avg_precision': avg_precision,
        'avg_recall': avg_recall,
        'avg_f1': avg_f1,
        'confusion_matrix': sum_cm
    }

# Função para plotar resultados
def plot_results(results_dict, filename):
    plt.figure(figsize=(15, 10))

    # Plot das métricas por fold
    plt.subplot(2, 2, 1)
    folds = range(1, 6)
    plt.plot(folds, results_dict['accuracies'], 'o-', label='Acurácia')
    plt.plot(folds, results_dict['precisions'], 's-', label='Precisão')
    plt.plot(folds, results_dict['recalls'], '^-', label='Recall')
    plt.plot(folds, results_dict['f1_scores'], 'd-', label='F1-Score')
    plt.xlabel('Fold')
    plt.ylabel('Valor')
    plt.title(f'Métricas por Fold - {os.path.basename(filename)}')
    plt.grid(True)
    plt.legend()

    # Barplot das médias das métricas
    plt.subplot(2, 2, 2)
    metrics = ['Acurácia', 'Precisão', 'Recall', 'F1-Score']
    values = [
        results_dict['avg_accuracy'],
        results_dict['avg_precision'],
        results_dict['avg_recall'],
        results_dict['avg_f1']
    ]
    plt.bar(metrics, values, color=['blue', 'green', 'red', 'purple'])
    plt.ylim(0, 1.0)
    for i, v in enumerate(values):
        plt.text(i, v + 0.02, f'{v:.4f}', ha='center')
    plt.title(f'Médias das Métricas - {os.path.basename(filename)}')

    # Matriz de confusão
    plt.subplot(2, 2, 3)
    if results_dict['confusion_matrix'].size > 0:
        sns.heatmap(
            results_dict['confusion_matrix'],
            annot=True,
            fmt='d',
            cmap='Blues',
            cbar=True
        )
        plt.title(f'Matriz de Confusão Acumulada - {os.path.basename(filename)}')
        plt.xlabel('Predito')
        plt.ylabel('Real')

    # Variabilidade das métricas (boxplot)
    plt.subplot(2, 2, 4)
    data_to_plot = [
        results_dict['accuracies'],
        results_dict['precisions'],
        results_dict['recalls'],
        results_dict['f1_scores']
    ]
    plt.boxplot(data_to_plot, labels=metrics)
    plt.ylim(0, 1.0)
    plt.title(f'Variabilidade das Métricas - {os.path.basename(filename)}')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()

    # Salvar a figura com nome do arquivo
    output_filename = f"resultados_{os.path.basename(filename).replace('.npy', '')}.png"
    plt.savefig(output_filename)
    print(f"Gráfico salvo como: {output_filename}")

    return plt

# Processar cada arquivo NPY
all_results = {}
all_datasets = []

# Primeiro passo: carregar todos os arquivos para determinar o tamanho da menor amostra
print("Carregando todos os arquivos para determinar tamanho mínimo de amostra...")
for npy_file in npy_files:
    # Carregar dados
    file_refs, embeds, labels, classes = load_embbed(npy_file)

    if len(embeds) == 0 or len(labels) == 0:
        print(f"Aviso: Dados insuficientes no arquivo {npy_file}. Pulando.")
        continue

    all_datasets.append({
        'file': npy_file,
        'file_refs': file_refs,
        'embeds': embeds,
        'labels': labels,
        'classes': classes,
        'size': len(embeds)
    })

# Determinar o tamanho da menor amostra entre todos os arquivos
if not all_datasets:
    print("Erro: Não foi possível carregar nenhum arquivo de dados válido.")
    exit()

min_dataset_size = min(dataset['size'] for dataset in all_datasets)
print(f"Tamanho da menor amostra entre todos os arquivos: {min_dataset_size}")

# Processar cada arquivo com o mesmo tamanho de amostra
for dataset in all_datasets:
    npy_file = dataset['file']
    embeds = dataset['embeds']
    labels = dataset['labels']
    classes = dataset['classes']

    print(f"\nProcessando: {npy_file}")
    print("=" * 70)

    print(f"Número original de exemplos: {len(embeds)}")
    print(f"Dimensão dos embeddings: {embeds.shape[1]}")
    print(f"Classes disponíveis: {classes}")

    # Verificar distribuição original de classes
    unique_labels, counts = np.unique(labels, return_counts=True)
    print("\nDistribuição original de classes:")
    for lbl, cnt in zip(unique_labels, counts):
        class_name = classes[lbl] if lbl < len(classes) else f"Classe {lbl}"
        print(f"  {class_name}: {cnt} exemplos")

    # Se este arquivo for maior que o menor, reduzir aleatoriamente preservando as proporções de classes
    if len(embeds) > min_dataset_size:
        print(f"\nReduzindo dataset para {min_dataset_size} amostras (tamanho da menor amostra)...")

        # Calcular proporção de cada classe
        proportions = counts / np.sum(counts)

        # Calcular quantas amostras manter de cada classe
        samples_per_class = np.round(proportions * min_dataset_size).astype(int)

        # Ajustar para garantir que o total seja exatamente min_dataset_size
        diff = min_dataset_size - np.sum(samples_per_class)
        if diff > 0:
            # Adicionar as amostras restantes à classe com maior proporção
            samples_per_class[np.argmax(proportions)] += diff
        elif diff < 0:
            # Remover as amostras excedentes da classe com maior proporção
            samples_per_class[np.argmax(proportions)] += diff

        # Selecionar amostras aleatoriamente de cada classe
        reduced_indices = []
        for i, label in enumerate(unique_labels):
            class_indices = np.where(labels == label)[0]
            np.random.seed(42 + i)  # Para reprodutibilidade, com seed diferente para cada classe
            selected_indices = np.random.choice(class_indices, samples_per_class[i], replace=False)
            reduced_indices.extend(selected_indices)

        # Reorganizar aleatoriamente os índices
        np.random.seed(42)
        np.random.shuffle(reduced_indices)

        # Reduzir o conjunto de dados
        embeds = embeds[reduced_indices]
        labels = labels[reduced_indices]

        # Verificar nova distribuição
        unique_labels_new, counts_new = np.unique(labels, return_counts=True)
        print("\nDistribuição após redução (mantendo proporção):")
        for lbl, cnt in zip(unique_labels_new, counts_new):
            class_name = classes[lbl] if lbl < len(classes) else f"Classe {lbl}"
            print(f"  {class_name}: {cnt} exemplos")

    # Treinar e avaliar o modelo
    results = train_and_evaluate(embeds, labels, classes)
    all_results[npy_file] = results

    # Mostrar resultados finais
    print("\nResultados finais:")
    print(f"Acurácia média: {results['avg_accuracy']:.4f}")
    print(f"Precisão média: {results['avg_precision']:.4f}")
    print(f"Recall médio: {results['avg_recall']:.4f}")
    print(f"F1-Score médio: {results['avg_f1']:.4f}")

    # Plotar resultados
    plot = plot_results(results, npy_file)
    plt.show()  # Mostrar o gráfico

    print("\n" + "=" * 70)

# Comparação entre os diferentes arquivos (se houver mais de um)
if len(all_results) > 1:
    plt.figure(figsize=(12, 8))

    # Comparar acurácias
    file_names = [os.path.basename(f).replace('.npy', '') for f in all_results.keys()]
    accuracies = [results['avg_accuracy'] for results in all_results.values()]
    precisions = [results['avg_precision'] for results in all_results.values()]
    recalls = [results['avg_recall'] for results in all_results.values()]
    f1_scores = [results['avg_f1'] for results in all_results.values()]

    x = np.arange(len(file_names))
    width = 0.2

    plt.bar(x - 1.5*width, accuracies, width, label='Acurácia', color='blue')
    plt.bar(x - 0.5*width, precisions, width, label='Precisão', color='green')
    plt.bar(x + 0.5*width, recalls, width, label='Recall', color='red')
    plt.bar(x + 1.5*width, f1_scores, width, label='F1-Score', color='purple')

    plt.xlabel('Arquivos')
    plt.ylabel('Valor Médio')
    plt.title('Comparação de Desempenho entre Arquivos')
    plt.xticks(x, file_names, rotation=45, ha='right')
    plt.ylim(0, 1.0)
    plt.legend()
    plt.tight_layout()

    # Salvar gráfico comparativo
    plt.savefig("comparacao_entre_arquivos.png")
    print("Gráfico comparativo salvo como: comparacao_entre_arquivos.png")
    plt.show()

print("Processamento concluído para todos os arquivos.")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Diretórios onde estão os embeddings
#directories = ["Data/terumo-data-jpeg", "Data/terumo-val"]
directories = ["Data/terumo-data-jpeg"]

# Cria uma lista de todos os arquivos .npy nos diretórios
npy_files = []
for dir_path in directories:
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.npy'):
            npy_files.append(os.path.join(dir_path, file_name))

# Função para carregar dados
def load_embbed(pickle_path: str) -> tuple:
    data = np.load(pickle_path, allow_pickle=True).tolist()
    file_refs = data['fnames']
    embeds = data['embeddings']
    labels = data['labels']
    classes = data['classes']
    embeds = np.vstack(embeds)
    labels = np.hstack(labels)
    return file_refs, embeds, labels, classes

# Função para treinar modelo com validação cruzada e gerar métricas
def train_and_evaluate(X, y, class_names):
    # Normalizar os dados
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Criar o modelo com camada escondida de 2049 neurônios
    model = MLPClassifier(
        hidden_layer_sizes=(2049,),  # Uma camada escondida com 2049 neurônios
        max_iter=1000,
        activation='relu',
        solver='adam',
        random_state=42
    )

    # Configurar validação cruzada com 5 folds
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    # Armazenar métricas para cada fold
    fold_accuracies = []
    fold_precisions = []
    fold_recalls = []
    fold_f1s = []
    fold_cms = []

    # Executar validação cruzada manualmente para capturar mais métricas
    fold_idx = 1
    for train_index, test_index in kf.split(X_scaled):
        X_train, X_test = X_scaled[train_index], X_scaled[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Treinar o modelo
        model.fit(X_train, y_train)

        # Fazer previsões
        y_pred = model.predict(X_test)

        # Calcular métricas
        accuracy = accuracy_score(y_test, y_pred)

        # Para multi-classe, calcular média ponderada
        precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        cm = confusion_matrix(y_test, y_pred)

        # Armazenar métricas
        fold_accuracies.append(accuracy)
        fold_precisions.append(precision)
        fold_recalls.append(recall)
        fold_f1s.append(f1)
        fold_cms.append(cm)

        print(f"Fold {fold_idx}:")
        print(f"  Acurácia: {accuracy:.4f}")
        print(f"  Precisão: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-Score: {f1:.4f}")
        print(classification_report(y_test, y_pred, target_names=class_names))
        print("-" * 50)

        fold_idx += 1

    # Calcular médias
    avg_accuracy = np.mean(fold_accuracies)
    avg_precision = np.mean(fold_precisions)
    avg_recall = np.mean(fold_recalls)
    avg_f1 = np.mean(fold_f1s)

    # Somar matrizes de confusão
    if fold_cms:
        sum_cm = fold_cms[0]
        for cm in fold_cms[1:]:
            sum_cm += cm
    else:
        sum_cm = np.array([])

    return {
        'accuracies': fold_accuracies,
        'precisions': fold_precisions,
        'recalls': fold_recalls,
        'f1_scores': fold_f1s,
        'avg_accuracy': avg_accuracy,
        'avg_precision': avg_precision,
        'avg_recall': avg_recall,
        'avg_f1': avg_f1,
        'confusion_matrix': sum_cm,
        'class_names': class_names  # Adicionado para usar na plotagem
    }

# Função para plotar resultados
def plot_results(results_dict, filename):
    plt.figure(figsize=(15, 10))

    # Plot das métricas por fold
    plt.subplot(2, 2, 1)
    folds = range(1, 6)
    plt.plot(folds, results_dict['accuracies'], 'o-', label='Acurácia')
    plt.plot(folds, results_dict['precisions'], 's-', label='Precisão')
    plt.plot(folds, results_dict['recalls'], '^-', label='Recall')
    plt.plot(folds, results_dict['f1_scores'], 'd-', label='F1-Score')
    plt.xlabel('Fold')
    plt.ylabel('Valor')
    plt.title(f'Métricas por Fold - {os.path.basename(filename)}')
    plt.grid(True)
    plt.legend()

    # Barplot das médias das métricas
    plt.subplot(2, 2, 2)
    metrics = ['Acurácia', 'Precisão', 'Recall', 'F1-Score']
    values = [
        results_dict['avg_accuracy'],
        results_dict['avg_precision'],
        results_dict['avg_recall'],
        results_dict['avg_f1']
    ]
    plt.bar(metrics, values, color=['blue', 'green', 'red', 'purple'])
    plt.ylim(0, 1.0)
    for i, v in enumerate(values):
        plt.text(i, v + 0.02, f'{v:.4f}', ha='center')
    plt.title(f'Médias das Métricas - {os.path.basename(filename)}')

    # Matriz de confusão com nomes das classes
    plt.subplot(2, 2, 3)
    if results_dict['confusion_matrix'].size > 0:
        class_names = results_dict['class_names']
        sns.heatmap(
            results_dict['confusion_matrix'],
            annot=True,
            fmt='d',
            cmap='Blues',
            cbar=True,
            xticklabels=class_names,
            yticklabels=class_names
        )
        plt.title(f'Matriz de Confusão Acumulada - {os.path.basename(filename)}')
        plt.xlabel('Predito')
        plt.ylabel('Real')

    # Variabilidade das métricas (boxplot)
    plt.subplot(2, 2, 4)
    data_to_plot = [
        results_dict['accuracies'],
        results_dict['precisions'],
        results_dict['recalls'],
        results_dict['f1_scores']
    ]
    plt.boxplot(data_to_plot, labels=metrics)
    plt.ylim(0, 1.0)
    plt.title(f'Variabilidade das Métricas - {os.path.basename(filename)}')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()

    # Salvar a figura com nome do arquivo
    output_filename = f"resultados_{os.path.basename(filename).replace('.npy', '')}.png"
    plt.savefig(output_filename)
    print(f"Gráfico salvo como: {output_filename}")

    return plt

# Processar cada arquivo NPY
all_results = {}

for npy_file in npy_files:
    print(f"\nProcessando: {npy_file}")
    print("=" * 70)

    # Carregar dados
    file_refs, embeds, labels, classes = load_embbed(npy_file)

    # Verificar se os dados foram carregados corretamente
    if len(embeds) == 0 or len(labels) == 0:
        print(f"Dados insuficientes no arquivo {npy_file}. Pulando.")
        continue

    print(f"Número de exemplos: {len(embeds)}")
    print(f"Dimensão dos embeddings: {embeds.shape[1]}")
    print(f"Classes disponíveis: {classes}")

    # Verificar distribuição de classes
    unique_labels, counts = np.unique(labels, return_counts=True)
    print("\nDistribuição de classes:")
    for lbl, cnt in zip(unique_labels, counts):
        class_name = classes[lbl] if lbl < len(classes) else f"Classe {lbl}"
        print(f"  {class_name}: {cnt} exemplos")

    # Treinar e avaliar o modelo
    results = train_and_evaluate(embeds, labels, classes)
    all_results[npy_file] = results

    # Mostrar resultados finais
    print("\nResultados finais:")
    print(f"Acurácia média: {results['avg_accuracy']:.4f}")
    print(f"Precisão média: {results['avg_precision']:.4f}")
    print(f"Recall médio: {results['avg_recall']:.4f}")
    print(f"F1-Score médio: {results['avg_f1']:.4f}")

    # Plotar resultados
    plot = plot_results(results, npy_file)
    plt.show()  # Mostrar o gráfico

    print("\n" + "=" * 70)

# Comparação entre os diferentes arquivos (se houver mais de um)
if len(all_results) > 1:
    plt.figure(figsize=(12, 8))

    # Comparar acurácias
    file_names = [os.path.basename(f).replace('.npy', '') for f in all_results.keys()]
    accuracies = [results['avg_accuracy'] for results in all_results.values()]
    precisions = [results['avg_precision'] for results in all_results.values()]
    recalls = [results['avg_recall'] for results in all_results.values()]
    f1_scores = [results['avg_f1'] for results in all_results.values()]

    x = np.arange(len(file_names))
    width = 0.2

    plt.bar(x - 1.5*width, accuracies, width, label='Acurácia', color='blue')
    plt.bar(x - 0.5*width, precisions, width, label='Precisão', color='green')
    plt.bar(x + 0.5*width, recalls, width, label='Recall', color='red')
    plt.bar(x + 1.5*width, f1_scores, width, label='F1-Score', color='purple')

    plt.xlabel('Arquivos')
    plt.ylabel('Valor Médio')
    plt.title('Comparação de Desempenho entre Arquivos')
    plt.xticks(x, file_names, rotation=45, ha='right')
    plt.ylim(0, 1.0)
    plt.legend()
    plt.tight_layout()

    # Salvar gráfico comparativo
    plt.savefig("comparacao_entre_arquivos.png")
    print("Gráfico comparativo salvo como: comparacao_entre_arquivos.png")
    plt.show()

print("Processamento concluído para todos os arquivos.")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import time
from tabulate import tabulate

# Diretórios onde estão os embeddings (apenas UNI.NPY)
uni_file_path = None
directories = ["Data/terumo-data-jpeg"]

# Busca o arquivo UNI.NPY
for dir_path in directories:
    for file_name in os.listdir(dir_path):
        if file_name.upper() == 'UNI.NPY':
            uni_file_path = os.path.join(dir_path, file_name)
            break
    if uni_file_path:
        break

if not uni_file_path:
    print("Arquivo UNI.NPY não encontrado nos diretórios especificados.")
    exit(1)

# Função para carregar dados
def load_embbed(pickle_path: str) -> tuple:
    data = np.load(pickle_path, allow_pickle=True).tolist()
    file_refs = data['fnames']
    embeds = data['embeddings']
    labels = data['labels']
    classes = data['classes']
    embeds = np.vstack(embeds)
    labels = np.hstack(labels)
    return file_refs, embeds, labels, classes

# Lista de classificadores para testar
def get_classifiers():
    classifiers = {
        "MLP": MLPClassifier(
            hidden_layer_sizes=(2049,),
            max_iter=1000,
            activation='relu',
            solver='adam',
            random_state=42
        ),
        "SVM": SVC(
            kernel='rbf',
            C=1.0,
            gamma='scale',
            random_state=42
        ),
        "Random Forest": RandomForestClassifier(
            n_estimators=100,
            max_depth=None,
            min_samples_split=2,
            random_state=42
        ),
        "KNN": KNeighborsClassifier(
            n_neighbors=5,
            weights='uniform',
            algorithm='auto'
        ),
        "Gradient Boosting": GradientBoostingClassifier(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=3,
            random_state=42
        )
    }
    return classifiers

# Função para treinar e avaliar um classificador com validação cruzada
def train_and_evaluate(X, y, class_names, classifier, classifier_name):
    print(f"\nTreinando e avaliando {classifier_name}...")

    # Normalizar os dados
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Configurar validação cruzada com 5 folds
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    # Armazenar métricas para cada fold
    fold_accuracies = []
    fold_precisions = []
    fold_recalls = []
    fold_f1s = []
    fold_cms = []
    fold_times = []

    # Executar validação cruzada manualmente para capturar mais métricas
    fold_idx = 1
    for train_index, test_index in kf.split(X_scaled):
        X_train, X_test = X_scaled[train_index], X_scaled[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Treinar o modelo e medir o tempo
        start_time = time.time()
        classifier.fit(X_train, y_train)
        train_time = time.time() - start_time

        # Fazer previsões
        y_pred = classifier.predict(X_test)

        # Calcular métricas
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        cm = confusion_matrix(y_test, y_pred)

        # Armazenar métricas
        fold_accuracies.append(accuracy)
        fold_precisions.append(precision)
        fold_recalls.append(recall)
        fold_f1s.append(f1)
        fold_cms.append(cm)
        fold_times.append(train_time)

        print(f"Fold {fold_idx} - Tempo de treino: {train_time:.2f}s")
        print(f"  Acurácia: {accuracy:.4f}")
        print(f"  Precisão: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-Score: {f1:.4f}")
        print("-" * 30)

        fold_idx += 1

    # Calcular médias
    avg_accuracy = np.mean(fold_accuracies)
    avg_precision = np.mean(fold_precisions)
    avg_recall = np.mean(fold_recalls)
    avg_f1 = np.mean(fold_f1s)
    avg_time = np.mean(fold_times)

    # Somar matrizes de confusão
    if fold_cms:
        sum_cm = fold_cms[0]
        for cm in fold_cms[1:]:
            sum_cm += cm
    else:
        sum_cm = np.array([])

    return {
        'classifier_name': classifier_name,
        'accuracies': fold_accuracies,
        'precisions': fold_precisions,
        'recalls': fold_recalls,
        'f1_scores': fold_f1s,
        'avg_accuracy': avg_accuracy,
        'avg_precision': avg_precision,
        'avg_recall': avg_recall,
        'avg_f1': avg_f1,
        'confusion_matrix': sum_cm,
        'class_names': class_names,
        'training_times': fold_times,
        'avg_training_time': avg_time
    }

# Função para plotar matriz de confusão
def plot_confusion_matrix(cm, class_names, classifier_name):
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        cbar=True,
        xticklabels=class_names,
        yticklabels=class_names
    )
    plt.title(f'Matriz de Confusão - {classifier_name}')
    plt.xlabel('Predito')
    plt.ylabel('Real')
    plt.tight_layout()
    plt.savefig(f"confusion_matrix_{classifier_name.replace(' ', '_').lower()}.png")
    plt.close()

# Função para plotar comparação de classificadores
def plot_classifier_comparison(all_results):
    # Preparar dados para gráficos
    classifier_names = [result['classifier_name'] for result in all_results]
    accuracies = [result['avg_accuracy'] for result in all_results]
    precisions = [result['avg_precision'] for result in all_results]
    recalls = [result['avg_recall'] for result in all_results]
    f1_scores = [result['avg_f1'] for result in all_results]
    train_times = [result['avg_training_time'] for result in all_results]

    # Criar figura com múltiplos subplots
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # Acurácia
    axes[0, 0].bar(classifier_names, accuracies, color='skyblue')
    axes[0, 0].set_title('Acurácia Média')
    axes[0, 0].set_ylim(0, 1.0)
    for i, v in enumerate(accuracies):
        axes[0, 0].text(i, v + 0.01, f'{v:.4f}', ha='center')

    # F1-Score
    axes[0, 1].bar(classifier_names, f1_scores, color='lightgreen')
    axes[0, 1].set_title('F1-Score Médio')
    axes[0, 1].set_ylim(0, 1.0)
    for i, v in enumerate(f1_scores):
        axes[0, 1].text(i, v + 0.01, f'{v:.4f}', ha='center')

    # Precisão e Recall
    x = np.arange(len(classifier_names))
    width = 0.35
    axes[1, 0].bar(x - width/2, precisions, width, label='Precisão', color='coral')
    axes[1, 0].bar(x + width/2, recalls, width, label='Recall', color='lightblue')
    axes[1, 0].set_title('Precisão e Recall Médios')
    axes[1, 0].set_xticks(x)
    axes[1, 0].set_xticklabels(classifier_names)
    axes[1, 0].set_ylim(0, 1.0)
    axes[1, 0].legend()

    # Tempo de Treinamento
    axes[1, 1].bar(classifier_names, train_times, color='plum')
    axes[1, 1].set_title('Tempo Médio de Treinamento (s)')
    for i, v in enumerate(train_times):
        axes[1, 1].text(i, v + 0.1, f'{v:.2f}s', ha='center')

    plt.tight_layout()
    plt.savefig("comparacao_classificadores.png")
    plt.close()

    # Gráfico de radar para comparação geral
    plt.figure(figsize=(10, 8))

    # Preparar dados para gráfico de radar
    metrics = ['Acurácia', 'Precisão', 'Recall', 'F1-Score']
    angles = np.linspace(0, 2*np.pi, len(metrics), endpoint=False).tolist()
    angles += angles[:1]  # Fechar o círculo

    # Adicionar eixos ao gráfico radar
    ax = plt.subplot(111, polar=True)
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_rlabel_position(0)
    plt.xticks(angles[:-1], metrics)

    # Plotar cada classificador
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    for i, result in enumerate(all_results):
        values = [
            result['avg_accuracy'],
            result['avg_precision'],
            result['avg_recall'],
            result['avg_f1']
        ]
        values += values[:1]  # Fechar o círculo
        ax.plot(angles, values, color=colors[i % len(colors)], linewidth=2, label=result['classifier_name'])
        ax.fill(angles, values, color=colors[i % len(colors)], alpha=0.1)

    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.title('Comparação de Desempenho dos Classificadores')
    plt.tight_layout()
    plt.savefig("radar_classificadores.png")
    plt.close()

    return fig

# Carregar dados do arquivo UNI.NPY
print(f"Carregando dados do arquivo: {uni_file_path}")
file_refs, embeds, labels, classes = load_embbed(uni_file_path)

# Verificar se os dados foram carregados corretamente
print(f"Número de exemplos: {len(embeds)}")
print(f"Dimensão dos embeddings: {embeds.shape[1]}")
print(f"Classes disponíveis: {classes}")

# Verificar distribuição de classes
unique_labels, counts = np.unique(labels, return_counts=True)
print("\nDistribuição de classes:")
for lbl, cnt in zip(unique_labels, counts):
    class_name = classes[lbl] if lbl < len(classes) else f"Classe {lbl}"
    print(f"  {class_name}: {cnt} exemplos")

# Obter classificadores para testar
classifiers = get_classifiers()

# Lista para armazenar resultados
all_results = []

# Treinar e avaliar cada classificador
for name, clf in classifiers.items():
    print(f"\n{'='*50}")
    print(f"Avaliando classificador: {name}")
    print(f"{'='*50}")

    results = train_and_evaluate(embeds, labels, classes, clf, name)
    all_results.append(results)

    # Mostrar resultados finais
    print("\nResultados finais:")
    print(f"Acurácia média: {results['avg_accuracy']:.4f}")
    print(f"Precisão média: {results['avg_precision']:.4f}")
    print(f"Recall médio: {results['avg_recall']:.4f}")
    print(f"F1-Score médio: {results['avg_f1']:.4f}")
    print(f"Tempo médio de treino: {results['avg_training_time']:.2f}s")

    # Plotar matriz de confusão para este classificador
    plot_confusion_matrix(results['confusion_matrix'], classes, name)

# Plotar comparação entre classificadores
plot_classifier_comparison(all_results)

# Encontrar o melhor classificador com base no F1-Score
best_classifier = max(all_results, key=lambda x: x['avg_f1'])
print("\n" + "="*70)
print(f"O melhor classificador é: {best_classifier['classifier_name']}")
print(f"F1-Score médio: {best_classifier['avg_f1']:.4f}")
print(f"Acurácia média: {best_classifier['avg_accuracy']:.4f}")
print("="*70)

# Exibir tabela de resultados
table_data = []
for result in all_results:
    table_data.append([
        result['classifier_name'],
        f"{result['avg_accuracy']:.4f}",
        f"{result['avg_precision']:.4f}",
        f"{result['avg_recall']:.4f}",
        f"{result['avg_f1']:.4f}",
        f"{result['avg_training_time']:.2f}s"
    ])

headers = ["Classificador", "Acurácia", "Precisão", "Recall", "F1-Score", "Tempo (s)"]
print("\nTabela comparativa dos classificadores:")
print(tabulate(table_data, headers=headers, tablefmt="grid"))
```
